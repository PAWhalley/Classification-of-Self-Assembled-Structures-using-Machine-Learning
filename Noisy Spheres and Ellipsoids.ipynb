<h1>Noisy Sphere

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
# set parameters
#sphere data
noise_parameter = 0.05 #amount of noise away from a true sphere.
m = 1000 #number of points.

x_eps = np.random.normal(size = m)*noise_parameter
y_eps = np.random.normal(size = m)*noise_parameter
z_eps = np.random.normal(size = m)*noise_parameter
# create dataset
Phi = 2*np.pi*np.random.rand(m) - np.pi
Theta = np.pi*np.random.rand(m) - 0.5*np.pi
X = (1+x_eps)*np.cos(Theta)*np.cos(Phi)
Y = (1+y_eps)*np.cos(Theta)*np.sin(Phi)
Z = (1+z_eps)*np.sin(Theta)

data_sphere = [X,Y,Z] #This would be an m by 3 array of coordinates which could be put into the neural network.

plt.plot(X,Y, 'x')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(X,Y,Z)

```

<h1>Noisy Ellipsoid


```python
noise_parameter = 0.05 #amount of noise away from a true sphere.
m = 1000 #number of points.
a_x = 1#x-direction stretch
b_y = 1#y-direction stretch
c_z  = 10#z-direction stretch
x_eps = np.random.normal(size = m)*noise_parameter
y_eps = np.random.normal(size = m)*noise_parameter
z_eps = np.random.normal(size = m)*noise_parameter
# create dataset
Phi = 2*np.pi*np.random.rand(m) - np.pi
Theta = np.pi*np.random.rand(m) - 0.5*np.pi
X = (a_x+x_eps)*np.cos(Theta)*np.cos(Phi)
Y = (b_y+y_eps)*np.cos(Theta)*np.sin(Phi)
Z = (c_z+z_eps)*np.sin(Theta)

data_ellipsoid = [X,Y,Z] #This would be an m by 3 array of coordinates which could be put into the neural network.

plt.plot(X,Y, 'x')
```
